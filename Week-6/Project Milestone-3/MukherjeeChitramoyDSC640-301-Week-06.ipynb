```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_excel('C:\\Users\\Chitramoy\\Desktop\\MS-DSC\\DSC-640\\Week-2\\nationaldatabaseofchildcareprices.xlsx')

# Preview the dataframe
df.head(5)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
 
# Aggregate data by StudyYear and calculate mean costs for each category
mean_costs = df.groupby('StudyYear')[['MCInfant', 'MCToddler', 'MCPreschool']].mean().reset_index()
 
# Plot the data
plt.figure(figsize=(14, 8))
 
plt.plot(mean_costs['StudyYear'], mean_costs['MCInfant'], marker='o', label='Infant Care')
plt.plot(mean_costs['StudyYear'], mean_costs['MCToddler'], marker='s', label='Toddler Care')
plt.plot(mean_costs['StudyYear'], mean_costs['MCPreschool'], marker='^', label='Preschool Care')
 
plt.xlabel('Year')
plt.ylabel('Average Cost')
plt.title('Increasing Cost of Childcare by Category')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Select necessary columns for MHI and MHI_2018
df_subset = df[['State_Name', 'MHI', 'MHI_2018']]

# Group by State_Name and take the mean of MHI and MHI_2018 (if multiple counties per state)
df_grouped = df_subset.groupby('State_Name').mean().reset_index()

# Select top 15 states based on MHI_2018
top_10_states = df_grouped.nlargest(15, 'MHI_2018')

# Plot the bar chart
plt.figure(figsize=(10, 6))
bar_width = 0.4
x = range(len(top_10_states))

# Plot bars for MHI and MHI_2018
plt.bar(x, top_10_states['MHI'], width=bar_width, label='MHI', align='center')
plt.bar([i + bar_width for i in x], top_10_states['MHI_2018'], width=bar_width, label='MHI_2018', align='center')

# Add labels and title
plt.xlabel('State')
plt.ylabel('Median Household Income')
plt.title('Comparison of MHI and MHI_2018 for Top 15 States')
plt.xticks([i + bar_width / 2 for i in x], top_10_states['State_Name'], rotation=45, ha='right')
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Select the columns of interest
df_subset = df[['EMP_M', 'EMP_Service', 'EMP_Sales', 'EMP_N', 'EMP_P']]

# Sum employment numbers across all counties for each category
employment_sums = df_subset.sum()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(employment_sums, labels=employment_sums.index, autopct='%1.1f%%', startangle=90)
plt.title('Employment Distribution in Affected Counties')
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.

# Display the plot
plt.show()
```

```python
# Select the columns of interest
df_subset = df[['StudyYear', 'FLFPR_20to64', 'FLFPR_20to64_Under6', 'FLFPR_20to64_6to17', 'FLFPR_20to64_Under6_6to17', 'MLFPR_20to64']]

# Aggregate data by StudyYear and calculate the mean for each category
mean_labor_force = df_subset.groupby('StudyYear').mean().reset_index()

# Plot the line graph
plt.figure(figsize=(14, 8))
plt.plot(mean_labor_force['StudyYear'], mean_labor_force['FLFPR_20to64'], marker='o', label='Female LFPR 20-64')
plt.plot(mean_labor_force['StudyYear'], mean_labor_force['FLFPR_20to64_Under6'], marker='s', label='Female LFPR 20-64 with Children <6')
plt.plot(mean_labor_force['StudyYear'], mean_labor_force['FLFPR_20to64_6to17'], marker='^', label='Female LFPR 20-64 with Children 6-17')
plt.plot(mean_labor_force['StudyYear'], mean_labor_force['FLFPR_20to64_Under6_6to17'], marker='d', label='Female LFPR 20-64 with Children <6 and 6-17')
plt.plot(mean_labor_force['StudyYear'], mean_labor_force['MLFPR_20to64'], marker='x', label='Male LFPR 20-64')

# Add labels, title, and legend
plt.xlabel('Year')
plt.ylabel('Labor Force Participation Rate')
plt.title('Labor Force Participation Rates Trends Over Years')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Aggregate data by StudyYear and calculate mean costs and median household income
income_data = df.groupby('StudyYear')[['MCBto5', 'MC6to11', 'MC12to17', 'MHI']].mean().reset_index()
 
# Plot the data
plt.figure(figsize=(14, 8))
plt.plot(income_data['StudyYear'], income_data['MCBto5'], marker='o', label='Childcare Cost for Bto5')
plt.plot(income_data['StudyYear'], income_data['MC6to11'], marker='s', label='Childcare Cost for 6to11')
plt.plot(income_data['StudyYear'], income_data['MC12to17'], marker='^', label='Childcare Cost for 12to17')
plt.plot(income_data['StudyYear'], income_data['MHI'], marker='x', label='Median Household Income')
plt.xlabel('Year')
plt.ylabel('Cost ($)')
plt.title('Childcare Costs and Median Household Income Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Aggregate data by StudyYear and calculate mean costs and economic indicators
economic_data = df.groupby('StudyYear')[['MCBto5', 'MC6to11', 'MC12to17', 'ME', 'FME', 'MME']].mean().reset_index()
 
# Plot the data
plt.figure(figsize=(14, 8))
plt.plot(economic_data['StudyYear'], economic_data['MCBto5'], marker='o', label='Childcare Cost for Bto5')
plt.plot(economic_data['StudyYear'], economic_data['MC6to11'], marker='s', label='Childcare Cost for 6to11')
plt.plot(economic_data['StudyYear'], economic_data['MC12to17'], marker='^', label='Childcare Cost for 12to17')
plt.plot(economic_data['StudyYear'], economic_data['ME'], marker='x', label='Median Earnings')
plt.plot(economic_data['StudyYear'], economic_data['FME'], marker='+', label='Female Median Earnings')
plt.plot(economic_data['StudyYear'], economic_data['MME'], marker='*', label='Male Median Earnings')
plt.xlabel('Year')
plt.ylabel('Cost and Earnings ($)')
plt.title('Childcare Costs and Economic Indicators Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Aggregate data by StudyYear and calculate mean costs and labor force participation rates
labor_force_data = df.groupby('StudyYear')[['MCBto5', 'MC6to11', 'MC12to17', 'FLFPR_20to64', 'MLFPR_20to64']].mean().reset_index()
 
# Plot the data
plt.figure(figsize=(14, 8))
plt.plot(labor_force_data['StudyYear'], labor_force_data['MCBto5'], marker='o', label='Childcare Cost for Bto5')
plt.plot(labor_force_data['StudyYear'], labor_force_data['MC6to11'], marker='s', label='Childcare Cost for 6to11')
plt.plot(labor_force_data['StudyYear'], labor_force_data['MC12to17'], marker='^', label='Childcare Cost for 12to17')
plt.plot(labor_force_data['StudyYear'], labor_force_data['FLFPR_20to64'], marker='x', label='Female Labor Force Participation Rate')
plt.plot(labor_force_data['StudyYear'], labor_force_data['MLFPR_20to64'], marker='+', label='Male Labor Force Participation Rate')
plt.xlabel('Year')
plt.ylabel('Cost and Participation Rate (%)')
plt.title('Childcare Costs and Labor Force Participation Rates Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Aggregate data by StudyYear and calculate mean costs and unemployment rates
unemployment_data = df.groupby('StudyYear')[['MCBto5', 'MC6to11', 'MC12to17', 'UNR_16', 'FUNR_16', 'MUNR_16']].mean().reset_index()
 
# Plot the data
plt.figure(figsize=(14, 8))
plt.plot(unemployment_data['StudyYear'], unemployment_data['MCBto5'], marker='o', label='Childcare Cost for Bto5')
plt.plot(unemployment_data['StudyYear'], unemployment_data['MC6to11'], marker='s', label='Childcare Cost for 6to11')
plt.plot(unemployment_data['StudyYear'], unemployment_data['MC12to17'], marker='^', label='Childcare Cost for 12to17')
plt.plot(unemployment_data['StudyYear'], unemployment_data['UNR_16'], marker='x', label='Unemployment Rate')
plt.plot(unemployment_data['StudyYear'], unemployment_data['FUNR_16'], marker='+', label='Female Unemployment Rate')
plt.plot(unemployment_data['StudyYear'], unemployment_data['MUNR_16'], marker='*', label='Male Unemployment Rate')
plt.xlabel('Year')
plt.ylabel('Cost and Unemployment Rate (%)')
plt.title('Childcare Costs and Unemployment Rates Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Demographic and Geographic Information:
    #1. State_Name: The name of the state where the childcare data was collected.
    #2. State_Abbreviation: Abbreviation for the state.
    #3. County_Name: The name of the county within the state.
    #4. County_FIPS_Code: The Federal Information Processing Standards (FIPS) code for the county.
    #5. StudyYear: The year when the data was collected.
    #6. TotalPop: Total population in the area.
    #7. OneRace, OneRace_W, OneRace_B, OneRace_I, OneRace_A, OneRace_H, OneRace_Other: Population breakdown by race.
    #8. Hispanic: Population of Hispanic ethnicity.
## Employment and Labor Force Participation: 
    #1. EMP_M, MEMP_M, FEMP_M, EMP_Service, MEMP_Service, FEMP_Service, EMP_Sales, MEMP_Sales, FEMP_Sales, EMP_N, MEMP_N, FEMP_N, EMP_P, MEMP_P, FEMP_P: Employment-         related variables.
    #2. FLFPR_20to64, FLFPR_20to64_Under6, FLFPR_20to64_6to17, FLFPR_20to64_Under6_6to17: Labor force participation rates for adults aged 20 to 64, categorized by           childcare responsibilities.
## Income and Household Information: 
    #1. MHI, ME, FME, MME: Median household income for different groups.
    #2. Households: Total number of households.
    #3. H_Under6_BothWork, H_Under6_FWork, H_Under6_MWork, H_Under6_SingleM: Households with children under 6 and their employment status.
## Childcare Costs: 
    #1. MCBto5, MC6to11, MC12to17, MC18to23, MC24to29, MC30to35, MC36to41, MC42to47, MC48to53, MC54toSA, MCSA: Childcare costs for different age groups.
    #2. MFCCBto5, MFCC6to11, MFCC12to17, MFCC18to23, MFCC24to29, MFCC30to35, MFCC36to41, MFCC42to47, MFCC48to53, MFCC54toSA, MFCCSA: Childcare costs adjusted for            family income.
## Flags for Childcare Costs:
    #1. Flags indicating whether childcare costs are available for infants, toddlers, and preschoolers.
